```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tqdm import tqdm

# Load Fashion MNIST dataset
(train_images, _), (_, _) = tf.keras.datasets.fashion_mnist.load_data()
train_images = train_images.astype(np.float32) / 255.0
train_images = train_images.reshape(train_images.shape[0], 784)

# Training Parameters
learning_rate = 0.0002
batch_size = 128
epochs = 100000
total_steps = 100000
print_interval = 2000

# Network Parameters
image_dim = 784
gen_hidden_dim = 256
disc_hidden_dim = 256
z_noise_dim = 100

# Xavier initialization
def xavier_init(shape):
    return tf.random.normal(shape=shape, stddev=1./tf.sqrt(shape[0]/2.0))

weights = {
    "disc_H": tf.Variable(xavier_init([image_dim, disc_hidden_dim])),
    "disc_final": tf.Variable(xavier_init([disc_hidden_dim, 1])),
    "gen_H": tf.Variable(xavier_init([z_noise_dim, gen_hidden_dim])),
    "gen_final": tf.Variable(xavier_init([gen_hidden_dim, image_dim]))
}

biases = {
    "disc_H": tf.Variable(xavier_init([disc_hidden_dim])),
    "disc_final": tf.Variable(xavier_init([1])),
    "gen_H": tf.Variable(xavier_init([gen_hidden_dim])),
    "gen_final": tf.Variable(xavier_init([image_dim]))
}

# Discriminator Network
def discriminator(x):
    hidden_layer = tf.nn.relu(tf.add(tf.matmul(x, weights["disc_H"]), biases["disc_H"]))
    final_layer = tf.add(tf.matmul(hidden_layer, weights["disc_final"]), biases["disc_final"])
    disc_output = tf.nn.sigmoid(final_layer)
    return final_layer, disc_output

# Generator Network
def generator(x):
    hidden_layer = tf.nn.relu(tf.add(tf.matmul(x, weights["gen_H"]), biases["gen_H"]))
    final_layer = tf.add(tf.matmul(hidden_layer, weights["gen_final"]), biases["gen_final"])
    gen_output = tf.nn.sigmoid(final_layer)
    return gen_output

# Losses
def discriminator_loss(real_output_disc, fake_output_disc):
    disc_real_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=real_output_disc, labels=tf.ones_like(real_output_disc)))
    disc_fake_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=fake_output_disc, labels=tf.zeros_like(fake_output_disc)))
    return disc_real_loss + disc_fake_loss

def generator_loss(fake_output_disc):
    return tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=fake_output_disc, labels=tf.ones_like(fake_output_disc)))

# Optimizers
discriminator_optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=learning_rate)
generator_optimizer = tf.keras.optimizers.legacy.Adam(learning_rate=learning_rate)

# Training
@tf.function
def train_step(x_batch, z_noise):
    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        _, disc_real_output = discriminator(x_batch)
        _, disc_fake_output = discriminator(generator(z_noise))
        
        d_loss = discriminator_loss(disc_real_output, disc_fake_output)
        g_loss = generator_loss(disc_fake_output)

    gradients_of_discriminator = disc_tape.gradient(d_loss, list(weights.values()))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, list(weights.values())))

    gradients_of_generator = gen_tape.gradient(g_loss, list(weights.values()))
    generator_optimizer.apply_gradients(zip(gradients_of_generator, list(weights.values())))

    return d_loss, g_loss

# Initialize the variables
num_samples_to_generate = 16
random_vector_for_generation = tf.random.normal([num_samples_to_generate, z_noise_dim])

# Training loop
for step in tqdm(range(total_steps + 1)):
    # Generate noise for generator input
    z_noise = np.random.uniform(-1., 1., size=[batch_size, z_noise_dim]).astype(np.float32)

    # Get a random batch of real images
    x_batch = train_images[np.random.randint(0, train_images.shape[0], size=batch_size)].reshape([batch_size, image_dim])

    # Update generator and discriminator
    d_loss, g_loss = train_step(x_batch, z_noise)

    # Print losses at specified intervals
    if step % print_interval == 0:
        print("Steps: {}: Generator Loss: {}, Discriminator Loss: {}".format(
            step, g_loss, d_loss))

    # Generate samples for display
    if step % print_interval == 0:
        generated_samples = generator(random_vector_for_generation)

# Display generated samples
n = int(np.sqrt(num_samples_to_generate))
canvas = np.empty((28 * n, 28 * n))
k = 0
for i in range(n):
    for j in range(n):
        canvas[i * 28:(i + 1) * 28, j * 28:(j + 1) * 28] = generated_samples[k].numpy().reshape(28, 28)
        k += 1

plt.figure(figsize=(n, n))
plt.imshow(canvas, origin="upper", cmap="gray")
plt.show()
```
